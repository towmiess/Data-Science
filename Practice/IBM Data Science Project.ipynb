```python
#TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year. (1 point) 

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Year': [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991],
        'Automobile_Sales': [100, 150, 120, 200, 250, 230, 180, 300, 280, 260, 220, 180]}

# Create a DataFrame using the sample data
df = pd.DataFrame(data)

# Create a figure and set the figure size
plt.figure(figsize=(10, 6))

# Group the data by 'Year' and calculate the mean of 'Automobile_Sales' for each year
df_line = df.groupby('Year')['Automobile_Sales'].mean()

# Plot a line chart using the mean sales data
df_line.plot(kind='line')

# Set custom ticks on the x-axis for all the years and rotate them for better visibility
plt.xticks(df['Year'], rotation=75)

# Set labels for the x and y axes
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.title('Automobile Sales over Time')

# Annotate years of recession on the plot
plt.annotate('1981-82 Recession', xy=(1982, 120), xytext=(1983, 180),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('1990 Recession', xy=(1990, 220), xytext=(1985, 250),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Task 1.2

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Vehicle_Type': ['Sedan', 'Sedan', 'SUV', 'SUV', 'Sedan', 'SUV'],
    'Sales': [150000, 160000, 170000, 140000, 180000, 190000]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Assuming you have information about recession years
recession_years = [2020, 2021]

# Create a line chart using pandas and matplotlib
plt.figure(figsize=(10, 6))

# Filter data for each vehicle type and plot lines
vehicle_types = df['Vehicle_Type'].unique()
for v_type in vehicle_types:
    v_type_data = df[df['Vehicle_Type'] == v_type]
    plt.plot(v_type_data['Year'], v_type_data['Sales'], marker='o', label=v_type)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Automobile Sales by Vehicle Type')

# Adding grid lines
plt.grid(True)

# Adding legend
plt.legend()

# Highlight recession periods
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Show the plot
plt.show()


```

```python
#TASK 1.4: To compare the variations in GDP during recession and non-recession periods using subplots
#you can use the matplotlib library to create a grid of subplots. Here's how you can do it:

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'GDP': [2000, 2100, 2200, 1900, 2300, 2400]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Assuming you have information about recession years
recession_years = [2020, 2021]

# Create subplots using matplotlib
fig, axs = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Plot for non-recession period
axs[0].plot(df['Year'], df['GDP'], marker='o', color='b')
axs[0].set_title('Non-Recession Period')
axs[0].set_ylabel('GDP')

# Plot for recession period
axs[1].plot(df['Year'], df['GDP'], marker='o', color='r')
axs[1].set_title('Recession Period')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('GDP')

# Highlight recession periods
for year in recession_years:
    axs[0].axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    axs[1].axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()

```

```python
#TASK 1.5: A bubble plot can effectively display the impact of seasonality on automobile sales by showing sales data with different bubble sizes representing sales values.
#Here's how you can create a bubble plot using matplotlib:

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1000, 1200, 1300, 1100, 1500, 1600, 1700, 1800, 1400, 1300, 1200, 1100]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Define bubble sizes based on sales values
bubble_sizes = df['Sales'] * 0.02  # Adjust the multiplier as needed

# Create a bubble plot using matplotlib
plt.figure(figsize=(10, 6))

# Scatter plot with bubble sizes and labels
plt.scatter(df.index, df['Sales'], s=bubble_sizes, alpha=0.7)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Impact of Seasonality on Automobile Sales')

# Customize x-axis labels
plt.xticks(df.index, df['Month'])

# Show the plot
plt.show()
```

```python
#TASK 1.6:To develop a scatter plot to identify the correlation between the average vehicle price and sales volume during recessions using Matplotlib
#you'll need to have a dataset with columns for 'Year', 'Average_Price', 'Sales', and information about recession years. Here's how you can create the scatter plot:

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Average_Price': [25000, 26000, 27000, 24000, 28000, 29000],
    'Sales': [150000, 160000, 170000, 140000, 180000, 190000]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Assuming you have information about recession years
recession_years = [2020, 2021]

# Create the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))

# Scatter plot for non-recession years
non_recession_years = [year for year in df['Year'] if year not in recession_years]
plt.scatter(df[df['Year'].isin(non_recession_years)]['Average_Price'],
            df[df['Year'].isin(non_recession_years)]['Sales'],
            color='b', label='Non-Recession Years', alpha=0.7)

# Scatter plot for recession years
plt.scatter(df[df['Year'].isin(recession_years)]['Average_Price'],
            df[df['Year'].isin(recession_years)]['Sales'],
            color='r', label='Recession Years', alpha=0.7)

# Adding labels and title
plt.xlabel('Average Vehicle Price')
plt.ylabel('Sales Volume')
plt.title('Correlation between Average Vehicle Price and Sales Volume')

# Adding legend
plt.legend()

# Highlight recession years
for year in recession_years:
    plt.axvline(x=df[df['Year'] == year]['Average_Price'].values[0],
                color='gray', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
# Install dash first if not already installed
!pip install dash

# Import necessary libraries
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Initialize the Dash app
app = dash.Dash(__name__)
app.title = "Interactive Dashboard for Sales Data"

# Sample data
df = pd.DataFrame({
    "Month": ["January", "February", "March", "April", "May", "June", "July", "August"],
    "Sales": [200, 150, 300, 400, 350, 450, 600, 550],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West"],
    "Product": ["A", "B", "A", "C", "B", "A", "C", "B"]
})

# Define the layout of the app with dropdowns
#Task 2.1: Develop a dashboard using Dash to display monthly sales data trends for different regions and products. 
app.layout = html.Div(children=[
    html.H1(children="Sales Data Dashboard", style={'textAlign': 'center'}),

    html.Div(children="This dashboard shows monthly sales data trends."),
    
#Task 2.2: Add dropdowns to the dashboard to allow users to select a region and a product to view sales trends for that specific region and product.
    html.Label('Select Region:'),
    dcc.Dropdown(
        id='region-dropdown',
        options=[
            {'label': 'North', 'value': 'North'},
            {'label': 'South', 'value': 'South'},
            {'label': 'East', 'value': 'East'},
            {'label': 'West', 'value': 'West'}
        ],
        value='North',  # Default value
        clearable=False
    ),

    html.Label('Select Product:'),
    dcc.Dropdown(
        id='product-dropdown',
        options=[
            {'label': 'Product A', 'value': 'A'},
            {'label': 'Product B', 'value': 'B'},
            {'label': 'Product C', 'value': 'C'}
        ],
        value='A',  # Default value
        clearable=False
    ),

    dcc.Graph(id='sales-graph')  # Graph placeholder
])

# Callback to update the graph based on dropdown selection
@app.callback(
    Output('sales-graph', 'figure'),
    [Input('region-dropdown', 'value'),
     Input('product-dropdown', 'value')]
)
def update_graph(selected_region, selected_product):
    # Filter the data based on the selected region and product
    filtered_df = df[(df['Region'] == selected_region) & (df['Product'] == selected_product)]

    # Create the line chart
    fig = px.line(filtered_df, x="Month", y="Sales", title=f"Sales Trends for {selected_product} in {selected_region}")

    return fig

# Run the Dash app
if __name__ == '__main__':
    app.run(debug=True)


```

```python
#Task 2.3
#install dash

!pip install dash

# Import necessary libraries
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Initialize the Dash app
app = dash.Dash(__name__)
app.title = "Interactive Dashboard for Sales Data"

# Sample data
df = pd.DataFrame({
    "Month": ["January", "February", "March", "April", "May", "June", "July", "August"],
    "Sales": [200, 150, 300, 400, 350, 450, 600, 550],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West"],
    "Product": ["A", "B", "A", "C", "B", "A", "C", "B"]
})

# Define the layout of the app with dropdowns and an output display
app.layout = html.Div(children=[
    html.H1(children="Sales Data Dashboard", style={'textAlign': 'center'}),

    html.Div(children="This dashboard shows monthly sales data trends."),

    html.Label('Select Region:'),
    dcc.Dropdown(
        id='region-dropdown',
        options=[
            {'label': 'North', 'value': 'North'},
            {'label': 'South', 'value': 'South'},
            {'label': 'East', 'value': 'East'},
            {'label': 'West', 'value': 'West'}
        ],
        value='North',  # Default value
        clearable=False
    ),

    html.Label('Select Product:'),
    dcc.Dropdown(
        id='product-dropdown',
        options=[
            {'label': 'Product A', 'value': 'A'},
            {'label': 'Product B', 'value': 'B'},
            {'label': 'Product C', 'value': 'C'}
        ],
        value='A',  # Default value
        clearable=False
    ),

    # Output Division with id and className
    html.Div(
        id='output-display',
        className='output-container',
        style={'margin-top': '20px', 'padding': '10px', 'border': '1px solid #ddd'}
    ),

    dcc.Graph(id='sales-graph')  # Graph placeholder
])

# Callback to update the graph and output display based on dropdown selection
@app.callback(
    [Output('sales-graph', 'figure'),
     Output('output-display', 'children')],
    [Input('region-dropdown', 'value'),
     Input('product-dropdown', 'value')]
)
def update_graph_and_output(selected_region, selected_product):
    # Filter the data based on the selected region and product
    filtered_df = df[(df['Region'] == selected_region) & (df['Product'] == selected_product)]

    # Create the line chart
    fig = px.line(filtered_df, x="Month", y="Sales", title=f"Sales Trends for {selected_product} in {selected_region}")

    # Create an output message
    output_message = f"Showing sales data for {selected_product} in the {selected_region} region."

    return fig, output_message

# Run the Dash app
if __name__ == '__main__':
    app.run(debug=True)
```
